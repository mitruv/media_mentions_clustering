## 1. Зробить кластеризацію даних за інфоприводами/наративами (можна аналізувати ЗМІ та соцмережі як разом, так і окремо, так і щось одне на вибір).

```python
from sentence_transformers import SentenceTransformer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
import re

def clean_text(text):
    text = str(text).lower()

    # Залишаємо тільки українські + англійські літери, цифри та пробіли
    text = re.sub(r'[^а-яіїєґa-z0-9\s]', ' ', text)

    # Зайві пробіли -> один
    text = re.sub(r'\s+', ' ', text).strip()
    
    text = re.sub(r'http\S+|www\.\S+', ' ', text)  # видалення посилань
    return text

df['текст'] = df['Заголовок'] + ' ' + df['Текст']
df['текст_очищений'] = df['текст'].apply(clean_text)
```

```python
model = SentenceTransformer('all-mpnet-base-v2')

# Перетворення тексту на вектори
embeddings = model.encode(df['текст_очищений'].tolist(), show_progress_bar=True)
```

```python
inertia = []
K = range(2, 11)

for k in K:
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    kmeans.fit(embeddings)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(K, inertia, marker='o')
plt.xlabel('Кількість кластерів')
plt.ylabel('WCSS (Inertia)')
plt.title('Метод ліктя для MPNet-ембедінгів')
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import silhouette_score

sil_scores = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(embeddings)
    sil = silhouette_score(embeddings, labels)
    sil_scores.append(sil)

plt.plot(range(2, 11), sil_scores, marker='x', color='red')
plt.title('Silhouette Score')
plt.xlabel('Кількість кластерів')
plt.grid(True)
plt.show()
```

```python
k = 6
kmeans = KMeans(n_clusters=k, random_state=42)
df['кластер'] = kmeans.fit_predict(embeddings)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(embeddings)

plt.figure(figsize=(10, 6))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=df['кластер'], cmap='Dark2')
plt.title('Кластеризація текстів через MPNet + PCA')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.colorbar(scatter, label='Кластер')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
stop_words_ua = set([
    'що', 'як', 'це', 'в', 'на', 'до', 'з', 'у', 'та', 'за', 'від', 'із', 'про',
    'але', 'тому', 'такий', 'ще', 'також', 'або', 'який', 'укра', 'україна', 'буде',
    'є', 'було', 'ми', 'ви', 'вони', 'ми', 'нас', 'нам', 'уже', 'вже', 'та', 'того'
])
```

```python
stop_words_ua.update([
    'день', 'дня', 'місяць', 'року', 'млн', 'тис', 'компанія', 'nestle',
    'даний', 'повідомляє', 'новина', 'український', 'українська', 'українці',
    'свій', 'власний', 'так', 'така', 'той', 'такий', 'бути', 'було', 'є'
])
```

```python
for i in range(df['кластер'].nunique()):
    cluster_text = ' '.join(df[df['кластер'] == i]['текст_очищений'])

    print(f"\nКластер {i}, кількість символів у тексті: {len(cluster_text)}")
    print("Перших 300 символів:\n", cluster_text[:300])
```

```python
from collections import Counter
import matplotlib.pyplot as plt

import re

# Функція для токенізації без стоп-слів
def tokenize(text):
    tokens = re.findall(r'\b[а-щґєіїюяА-ЩҐЄІЇЮЯa-zA-Z]{3,}\b', text.lower())
    return [token for token in tokens if token not in stop_words_ua]

# Візуалізація частотних слів по кожному кластеру
for i in range(df['кластер'].nunique()):
    cluster_text = ' '.join(df[df['кластер'] == i]['текст_очищений'])
    words = tokenize(cluster_text)
    word_freq = Counter(words).most_common(10)

    if word_freq:
        words, counts = zip(*word_freq)
        plt.figure(figsize=(6, 3))
        plt.bar(words, counts)
        plt.title(f'🔠 Топ-слова у кластері {i}')
        plt.ylabel("Частота")
        plt.xticks(rotation=45)
        plt.grid(True)
        plt.tight_layout()
        plt.show()
    else:
        print(f"Кластер {i}: немає слів після фільтрації.")
```
